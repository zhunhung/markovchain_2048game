```python
import copy
import itertools

a=[2,2]

all_states = [[0],[2,2]]

def generate_states(state):
    if state[-1]==2048:
        return
    else:
        state.sort()
        if len(state) == len(set(state)):
            pass
        else:
            result = []
            uniqueEle = list(set(state))
            for ele in uniqueEle:
                if state.count(ele) == 2:
                    result.append(ele*2)
                elif state.count(ele) == 3:
                    result.append(ele*2)
                    result.append(ele)
                else:
                    result.append(ele)
            result.sort()
            state = result


        new_state_2 = copy.deepcopy(state)
        new_state_4 = copy.deepcopy(state)
        new_state_2.insert(0, 2)
        new_state_2.sort()
        new_state_4.insert(0, 4)
        new_state_4.sort()
        if(new_state_2 not in all_states and new_state_4 not in all_states):
            all_states.append(new_state_2)
            all_states.append(new_state_4)
            generate_states(new_state_2)
            generate_states(new_state_4)
        elif (new_state_2 not in all_states and new_state_4 in all_states):
            all_states.append(new_state_2)
            all_states.append(new_state_4)
            generate_states(new_state_2)
        elif (new_state_2 in all_states and new_state_4 not in all_states):
            all_states.append(new_state_2)
            all_states.append(new_state_4)
            generate_states(new_state_4)
        else:
            all_states.append(new_state_2)
            all_states.append(new_state_4)
            return
        

generate_states(a)

all_states.sort()
all_states = sorted(all_states, key=sum)
final_list = list(all_states for all_states,_ in itertools.groupby(all_states))
print(final_list)
print(len(final_list))
```

```python
final_list.sort()
```

```python
final_list_str = [str(row) for row in final_list]
```

```python
sumArr = [sum(row) for row in final_list]
```

```python
lenArr = [len(row) for row in final_list]
lenArr
```

```python
import pandas as pd
df = pd.DataFrame({'state':final_list_str,'len':lenArr, 'sum':sumArr})
df
```

```python
#Sort and save csv
newdf = df.sort_values(by=['sum','len']).reset_index(drop=True)
newdf.to_csv('generated_states.csv', index=False)
```
